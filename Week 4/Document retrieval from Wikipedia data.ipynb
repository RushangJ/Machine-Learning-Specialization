```python
import turicreate as tc
```

```python
# Load text data- from wikipidia,pages on pepole
pepole=tc.SFrame('datasets/pepoledata')
```

```python
pepole
```

```python
len(pepole)
```

```python
# Explore the dataset and checkout the text it contains
obama=pepole[pepole['name']=='Barack Obama']
obama['text']
```

```python
clooney= pepole[pepole['name']== 'George Clooney']
clooney['text']
```

```python
# Get the word count for Obama article
obama['word_count']=tc.text_analytics.count_words(obama['text'])
```

```python
obama
```

```python
obama['word_count']
```

```python
# Sort the word count for the obama article
obama_word_count_table= obama[['word_count']].stack('word_count',new_column_name=['word','count'])
```

```python
obama_word_count_table
```

```python
obama_word_count_table.sort('count',ascending=False)
```

```python
# Comput TF-IDF for the corpus
pepole['word_count']= tc.text_analytics.count_words(pepole['text'])
pepole.head()
```

```python
pepole['tfidf']= tc.text_analytics.tf_idf(pepole['text'])
pepole
```

```python
pepole['tfidf']
```

```python
# Examine the TF-IDF for the obama article
obama= pepole[pepole['name']=='Barack Obama']
```

```python
obama[['tfidf']].stack('tfidf',new_column_name=['word','tfidf']).sort('tfidf',ascending=False)
```

```python
# Manually compute distances between a few pepole
clinton= pepole[pepole['name']=='Bill Clinton']
```

```python
beckham= pepole[pepole['name']=='David Beckham']
```

```python
# Is Obama closer to Clinton than to Beckham?
tc.distances.cosine(obama['tfidf'][0],clinton['tfidf'][0])
```

```python
tc.distances.cosine(obama['tfidf'][0],beckham['tfidf'][0])
```

```python
# Build a nearest neighbor model for document retrieval
knn_model= tc.nearest_neighbors.create(pepole,features=['tfidf'],label='name')
```

```python
# Applying the nearest-neighbors model for retrieval
# Who is closest to Obama?
knn_model.query(obama)
```

```python
# Other examples of document retrieval
swift= pepole[pepole['name']=='Taylor Swift']
```

```python
knn_model.query(swift)
```

```python
jolie= pepole[pepole['name']=='Angelina Jolie']
```

```python
knn_model.query(jolie)
```

```python
arnold= pepole[pepole['name']=='Arnold Schwarzenegger']
```

```python
knn_model.query(arnold)
```

```python
# Whats the most similar article, other than itself, to the one on Elton John using word count features?
john= pepole[pepole['name']=='Elton John']
```

```python
john[['word_count']].stack('word_count', new_column_name = ['word','count']).sort('count',ascending=False)
```

```python
john[['tfidf']].stack('tfidf', new_column_name = ['word','tfidf']).sort('tfidf',ascending=False)
```

```python
beckham = pepole[pepole['name'] == 'Victoria Beckham']
```

```python
tc.distances.cosine(john['tfidf'][0],beckham['tfidf'][0])
```

```python
mccartney = pepole[pepole['name'] == 'Paul McCartney']
```

```python
tc.distances.cosine(john['tfidf'][0], mccartney['tfidf'][0])
```

```python
word_count_cosine_model = tc.nearest_neighbors.create(pepole,features=['word_count'],label='name',distance='cosine')
```

```python
word_count_cosine_model.query(john)
```
